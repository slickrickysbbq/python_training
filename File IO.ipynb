```python
import os

# We are going to create a directory under the Works directory where we will perform
# some file I/O operations and demonstrate other functionality offered by os
# and sub-modules

lab_dir = './lab_dir'
if not(os.path.exists(lab_dir)):
    os.mkdir(lab_dir)
    
# list the contents of the directory
os.listdir(lab_dir)

# Build the path to the file to be created
file_name = 'os_dir.txt'
full_file_name = os.path.join(lab_dir, file_name); full_file_name
# Note: Python escapes the Windows OS backslash with an additional backslash
```
